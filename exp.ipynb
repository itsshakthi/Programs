```python
#Position of substring 
a=input()
b=input()

l=len(a)
l2=len(b)
for i in range(l-l2+1):
    if a[i:i+l]==b:
        print(i+1)
        break
else:
    print("Wrong")
```

```python
n=int(input())
for i in range(n):
    for j in range(n-i-1):
        print(' ',end=' ')
    for j in range(i+1):
        print(j+1,end=' ')
    if i>0:
        for m in range(i):
            print(i-m,end=' ')
    print()
```

```python
n=int(input())
for i in range(1,n+1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        print('a',end=' ')
    print()
for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        print('a',end=' ')
    print()
```

```python
a=int(input())
l=[int(s) for s in range(1,a+1)]
l1=l[::-1]
for i in range(1,a+1):
    for j in range(a-i):
        print(" ",end=" ")
    print(*l[:i]+l1[a-i+1:])
```
