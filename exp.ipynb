```python
n=int(input())
for i in range(n):
    for j in range(n-i-1):
        print(' ',end=' ')
    for j in range(i+1):
        print(j+1,end=' ')
    if i>0:
        for m in range(i):
            print(i-m,end=' ')
    print()
```

```python
n=int(input())
for i in range(1,n+1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        print('a',end=' ')
    print()
for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        print('a',end=' ')
    print()
```

```python
s=input()
l=[]
d=[]
n,i=len(s),0
while i<n:
    a=''
    if s[i].isnumeric():
        while i<n and s[i].isnumeric():
            a=a+s[i]
            i+=1
        l.append(a)
    else:
        while i<n and s[i].isalpha():
            a=a+s[i]
            i+=1
        d.append(a)
print(l,d)
```

```python
#Printing symbol in most repeating char
from collections import Counter
s=input()
c=Counter(s)
m=max(c.values())
l=None
for i in c:
    if c[i]==m:
        l=i
g=''
for i in s:
    if i==l:
        g=g+'*'
    else:
        g=g+i
print(g)
```
